```python
import pandas as pd
import numpy as np
```

```python
def gradient_descent(x,y):
    m_curr = b_curr = 0
    iterations = 1000
    n=len(x)
    learning_rate = 1
    
    for i in range(iterations):
        y_predicted = m_curr*x + b_curr
        cost_func = (1/n)*sum([val**2 for val in (y-y_predicted)])
        m_derivatives = -(2/n)*sum(y-y_predicted)
        m_curr = m_curr - learning_rate*m_derivatives
        b_derivatives = -(2/n)*sum(y-y_predicted)
        b_curr = b_curr - learning_rate*b_derivatives
        
        print('m {}, b {}, cost {}, iteration {}'.format(m_curr,b_curr,cost_func,i))
```

```python
df=pd.read_csv('covid_19_dataset.csv')
df.columns
```

```python
df=df.fillna(method='ffill') 
new_df=df.dropna()
new_df
```

```python
x=new_df['total_cases']
x
```

```python
y=new_df['total_deaths']
y
```

```python
gradient_descent(x,y)
```
